###**EXERCISE 12**: ***



Plot required data as shown below:
>$y_1=x^3*e^{-x^2}$

>$y_2=x^4*e^{-x^2}$

>$y_3=x^5*e^{-x^2}$


```
from math import e
x = np.arange(0., 3., 0.05)
r = x**3*e**(-x**2)
s = x**4*e**(-x**2)
t = x**5*e**(-x**2)
```

```
from matplotlib.pyplot import figure

figure(figsize=(10.5, 5), dpi=80)

plt.plot(x, r, ls='-.', c = 'b', label='$y_1=x^3*e^{-x^2}$');
plt.plot(x, s, ds='steps', c = 'c', label='$y_2=x^4*e^{-x^2}$');
plt.plot(x, t, ls='None', marker='^', c = '#99A1CD', label='$y_3=x^5*e^{-x^2}$');
plt.grid(color='g', linestyle='-', linewidth=.5, alpha=.1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('MATPLOTLIB PLOTTING')
plt.legend(loc=(1.02,0));
```

###**EXERCISE 16**: ***



Plot the following axeses with provided data:

```
x = np.arange(1,5,0.1)

t = x**3
s = x**5*np.exp(-x**2)
u = np.sin(x)
v = np.sqrt(x)
```

```
fig, axs = plt.subplots(1,2, figsize=(15, 6));

axs[0].plot(x,t,marker='o',c='k',markerfacecolor='c',label='t');
axs[1].plot(x,s,marker='o',c='k',markerfacecolor='y',label='s');

ax1 = fig.add_axes([0.2, 0.6, 0.15, 0.2]);
ax1.plot(x,v,ls='--',c='r',label='v')

ax2 = fig.add_axes([0.7, 0.6, 0.15, 0.2]);
ax2.plot(x,u,ls='--',c='g',label='u')

axs[0].set_xlabel('x')
axs[1].set_xlabel('x')
ax1.set_xlabel('x')
ax2.set_xlabel('x')
axs[0].set_ylabel('t')
axs[1].set_ylabel('s')
ax1.set_ylabel('v')
ax2.set_ylabel('u')

fig.legend(loc=5);
```

###**EXERCISE 18**: ****



Using the data set available at [https://raw.githubusercontent.com/immune-institute/viz-datasets/main/wines_reviews.csv](https://raw.githubusercontent.com/immune-institute/viz-datasets/main/wines_reviews.csv) create the bar plot shown below:

```
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/immune-institute/viz-datasets/main/wines_reviews.csv")

df.sample(10)
```

```
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 9));

df.groupby('points')['Unnamed: 0'].nunique().plot(kind='bar', ax=axes[0,0], color='m', title='Wine Scores', xlabel="", ylabel="");
df[['price']].plot(kind='hist',bins=[0,500,1000,1500,2000,2500,3000], rwidth=0.5, ax=axes[0,1], color='m', legend=False, title= 'Wine Prices', xlabel="", ylabel="");
df.groupby('variety')['Unnamed: 0'].nunique().sort_values(ascending=False)[:20].plot(kind='bar', width=.8, ax=axes[1,0], color='m', title= 'Wine Varieties', xlabel="", ylabel="");
df.groupby('province')['Unnamed: 0'].nunique().sort_values(ascending=False)[:20].plot(kind='bar', ax=axes[1,1], color='m', title= 'Wine Origins', xlabel="", ylabel="");
```
